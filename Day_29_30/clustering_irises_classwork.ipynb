```python
# Clustering in machine learning

# what is clustering?

# Clustering is a type of unsupervised learning. It is used to find groups in data.

# there are many clustering algorithms, but the most common is k-means clustering.
```

```python
# we will use scikit-learn to perform k-means clustering on the iris dataset.

# import the libraries we will need
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('../data/iris-data-clean.csv')
df.head()
```

```python
df.describe()
```

```python
# draw histogram using seaborn
import seaborn as sns
sns.set()
df.hist(bins=30, figsize=(20,15))
```

```python
# extract features into numpy array
features = df[['sepal_length_cm', 'sepal_width_cm', 'petal_length_cm', 'petal_width_cm']].values
features[:5]
```

```python
# for clustering we are not going to split into train and test sets
# again this is unsupervised learning
```

```python
# the most popular clustering algorithm is k-means clustering
# we will use scikit-learn to perform k-means clustering

# create a k-means object
kmeans = KMeans(n_clusters=3, random_state=0)
# we cheat a litle since we know there are 3 species of iris
```

```python
type(kmeans)
```

```python
# pefrom k-means clustering
kmeans.fit(features) # this is where the magic happens

```

```python
# show the results
print(kmeans.labels_)
```

```python
# so we have 3 clusters (0, 1, 2) - arbitrary labels
# we can already see that it is not perfect but it is a good start
```

```python
# let's compare results with the actual species
print(df['class'].values)
```

```python
# create numerical labels for the species and save in a species column
df['species'] = df['class'].map({'Iris-setosa': 0, 'Iris-versicolor': 2, 'Iris-virginica': 1})
df.head()
```

```python
df.tail()
```

```python
# now compare species and labels
# print the number of matches
print(sum(df['species'] == kmeans.labels_))
```

```python
# thus our accuracy is 134 / 149
print(sum(df['species'] == kmeans.labels_) / len(df)) # so roughly 90% accuracy
```

```python
# plot the results using matplotlib
# first two features and color by labels
plt.scatter(features[:,0], features[:,1], c=kmeans.labels_)
```

```python
# so k-means came up with a new species of irises :)
# so in real life we would not know the number of clusters
```

```python
# with kmeans we tell the algorithm how many clusters to find
# but what if we don't know how many clusters there are?

# we can use the elbow method to find the optimal number of clusters # TODO look up elbow method
```

```python
# we will use the inertia_ attribute to find the optimal number of clusters
# inertia is the sum of squared distances of samples to their closest cluster center
# so the lower the inertia the better

# create a list of inertia values
# we can use scikit-learn metrics module to calculate the inertia

from sklearn.metrics import pairwise_distances_argmin_min
```

```python
# we will test kmeans with 2 to 10 clusters and calculate the inertia
inertia = []
MAX_CLUSTERS = 20 + 1
for k in range(2, MAX_CLUSTERS):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(features)
    inertia.append(kmeans.inertia_)
    # so we throw away the labels and keep the inertia

# kmeans is fast and we have little data so we can do this
# but in real life some algorithms are slow and we have a lot of data which would make this unfeasible

# plot the inertia values
plt.plot(range(2, MAX_CLUSTERS), inertia, marker='o')
```

```python
# now if we our cluster size was 149 we would have a perfect fit but use would be useless
# everyone would have their own cluster
# also cluster size 1 is similarly useless

# helpful links on elbow method
# https://www.geeksforgeeks.org/elbow-method-for-optimal-value-of-k-in-kmeans/
```

```python
# let's try another clustering algorithm

# we will use DBSCAN

# DBSCAN is a density based clustering algorithm

# in DBSCAN we specify two parameters
# eps - the maximum distance between two samples for them to be considered as in the same neighborhood
# min_samples - the number of samples in a neighborhood for a point to be considered as a core point

# DBSCAN will find all samples that are in the same neighborhood as a core point

# DBSCAN will not find clusters that are not dense enough
# so it will not find clusters that are separated by a large distance

# let's create a DBSCAN object
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.5, min_samples=5)
```

```python
# again features is a 2d numpy array with 4 features in our case
# you could use a list of lists instead but it would be slower
features[:5]
```

```python
# let's fit the model
dbscan.fit(features)
# print the labels
print(dbscan.labels_)
```

```python
# -1 in DBSCAN means that the point is considered an outlier
```

```python
# let's plot the results
plt.scatter(features[:,0], features[:,1], c=dbscan.labels_)
```

```python
# let's plot 3D plot of first 3 features and color by labels
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(features[:,0], features[:,1], features[:,2], c=dbscan.labels_)
plt.show()
```

```python
# as the number of dimensions increases it becomes harder to visualize the data
# also we encounter something called the curse of dimensionality

# curse of dimensionality is the fact that as the number of dimensions increases
# the amount of data needed to represent the data increases exponentially
# also the distance between points increases exponentially
# more about curse of dimensionality here
# https://en.wikipedia.org/wiki/Curse_of_dimensionality
```

```python
# how do we deal with the curse of dimensionality?
# we can use dimensionality reduction techniques
# we can use PCA to reduce the number of dimensions - TODO look up PCA
# we can use t-SNE to reduce the number of dimensions
```

```python
# PCA stands for principal component analysis
# it is a linear dimensionality reduction technique
# it finds the directions of maximum variance in high-dimensional data and projects it onto a smaller dimensional subspace while retaining most of the variance
# link on how PCA works
# https://towardsdatascience.com/a-one-stop-shop-for-principal-component-analysis-5582fb7e0a9c

# fantastic explanation on PCA is here:
# https://stats.stackexchange.com/questions/2691/making-sense-of-principal-component-analysis-eigenvectors-eigenvalues
```

```python
# lets reduce our features to 2 dimensions using PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(features) # notice again fit is a method not an attribute
# the method which does all the work is fit as usual

# show the transformed features - you do need transform as well
features_2d = pca.transform(features)

# you could do this in one step
# features_2d = PCA(n_components=2).fit_transform(features)

# print first 5 rows of features_2d
features_2d[:5] # these features have no special meaning - they are just the first 2 principal components
```

```python
# plot the results - we can see that the clusters are more separated
plt.scatter(features_2d[:,0], features_2d[:,1], c=dbscan.labels_)
```

```python
# lets retrain the DBSCAN model with the new features
dbscan.fit(features_2d)
# print the labels
print(dbscan.labels_)
```

```python
# plot the results
plt.scatter(features_2d[:,0], features_2d[:,1], c=dbscan.labels_)
```

```python
# lets plot kmeans elbow method with the new features
inertia = []
MAX_CLUSTERS = 20 + 1
for k in range(2, MAX_CLUSTERS):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(features_2d)
    inertia.append(kmeans.inertia_)
    # so we throw away the labels and keep the inertia

# kmeans is fast and we have little data so we can do this
# but in real life some algorithms are slow and we have a lot of data which would make this unfeasible

# plot the inertia values
plt.plot(range(2, MAX_CLUSTERS), inertia, marker='o')
```

```python
# lets use kmeans 3 clusters again
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(features_2d)

# plot the results
plt.scatter(features_2d[:,0], features_2d[:,1], c=kmeans.labels_)

# features_2d[:,0] means all rows of the first column
# features_2d[:,1] means all rows of the second column
```
